```python
# Matplotlib is a comprehensive library 
# for creating static, animated, and interactive visualizations in Python.

#importing the library as shortname 
import matplotlib.pyplot as plt 


# for later 
import numpy as np
```

```python
# simple line plot 
plt.plot([1,1])
```

```python
# figure - It is a diagram or a shape that can be formed by a collection of plots in different dimensions.
# plot(x, y) # plot x and y using default line style and color
plt.figure(1)
plt.plot([1,1])
plt.figure(2)
plt.plot([1,2])


# Figure(1) helps print the first graph with plot([1,1]), and figure(2) helps print the second graph with plot([1,2]).
```

```python
# adding labels and titles 

# plt.plot(*args, scalex=True, scaley=True, data=None, **kwargs)
plt.plot([1,2,1,2])
plt.title('TITLE OF CHART')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
```

```python
# example line plot 

x = [1,2,3,4,5]
y = [50,40,70,80,20]
y2 = [80,20,20,50,60]
y3 = [70,20,60,40,60]
y4 = [80,20,20,50,60]

#plot takes a bunch of arguments 
# read documentation here: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html
plt.plot(x,y,'g',label='Enfield', linewidth=5)
plt.plot(x,y2,'c',label='Honda',linewidth=5)
plt.plot(x,y3,'k',label='Yahama',linewidth=5)
plt.plot(x,y4,'y',label='KTM',linewidth=5)
plt.title('bike details in line plot')
plt.ylabel('Distance in kms')
plt.xlabel('Days')
# adds a legend 
plt.legend()
```

```python
# example bar chart 

import matplotlib.pyplot as plt

# bar (list of x and y)
# if you don't set something like the color, it will use the default color 
# in this case, it is blue 
# color can someone take letters, names, or hex, rgba 

plt.bar([0.25,1.25,2.25,3.25,4.25],[50,40,70,80,20],
label='Enfield',width=.5)
plt.bar([0.26,1.25,2.25,3.25,4.25],[80,20,20,50,60],
label='Honda', color='r',width=.5)
plt.bar([0.31,1.5,2.5,3.5,4.5],[70,20,60,40,60],
label='Yamaha', color='y',width=.5)
plt.bar([.75,1.75,2.75,3.75,4.75],[80,20,20,50,60],
label='KTM', color='g',width=.5)
plt.legend()
plt.xlabel('Days')
plt.ylabel('Distance (kms)')
plt.title('Bikes details in BAR PLOTTING')
```

```python
# Area plot 

import matplotlib.pyplot as plt
days = [1,2,3,4,5]
Enfield =[50,40,70,80,20]
Honda = [80,20,20,50,60]
Yahama =[70,20,60,40,60]
KTM = [80,20,20,50,60]

# plotting empty arrays 
plt.plot([],[],color='k', label='Enfield', linewidth=5)
plt.plot([],[],color='c', label='Honda', linewidth=5)
plt.plot([],[],color='y', label='Yahama', linewidth=5)
plt.plot([],[],color='m', label='KTM', linewidth=5)

# stacking using the arrays assigned above 
plt.stackplot(days, Enfield, Honda, Yahama, KTM, colors=['k','c','y','m'])
plt.xlabel('Days')
plt.ylabel('Distance in kms')
plt.title('Bikes details in area plot')
plt.legend()
```

```python
# pie plot 

import matplotlib.pyplot as plt
days = [1,2,3,4,5]
Enfield =[50,40,70,80,20]
Honda = [80,20,20,50,60]
Yahama =[70,20,60,40,60]
KTM = [80,20,20,50,60]
slices = [8,5,5,6]
activities = ['Enfield','Honda','Yahama','KTM']
cols = ['c','g','y','b']
plt.pie(slices,
labels=activities,
colors=cols,
startangle=90,
shadow= True,
#To “explode” a pie chart means to make one of the wedges of the pie chart to stand out. 
explode=(0,0.1,0,0),
autopct='%1.1f%%') # to the nearest .1% float 
plt.title('Bike details in Pie Plot')
```

```python
# scatter plot 

import matplotlib.pyplot as plt
days = [1, 2, 3, 4, 5]
Y1 = [50, 40, 70, 80, 20]
Y2=[80, 20, 20, 50, 60]
Y3=[70, 20, 60, 40, 60]
Y4=[80, 20, 20, 50, 60]
plt.scatter(days,Y1, label='Enfield',color='r')
plt.scatter(days,Y2,label='Honda',color='b')
plt.scatter(days,Y3,label='Yahama',color='y')
plt.scatter(days,Y4,label='KTM',color='k')
plt.xlabel('Days')
plt.ylabel('Distance in kms')
plt.title('Bike details in Scatter Plot')
plt.legend()
```

```python
#3d Plot 

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()

#argsint, (int, int, index), or SubplotSpec, default: (1, 1, 1)
#Three integers (nrows, ncols, index). The subplot will take the index position on a grid with nrows rows and ncols columns

ax = fig.add_subplot(111, projection='3d')
x = [1,2,3,4,5]
y = [50,40,70,80,20]
y2 = [80,20,20,50,60]
y3 = [70,20,60,40,60]
y4 = [80,20,20,50,60]
plt.plot(x,y,'g',label='Enfield', linewidth=5)
plt.plot(x,y2,'c',label='Honda',linewidth=5)
plt.plot(x,y3,'k',label='Yahama',linewidth=5)
plt.plot(x,y4,'y',label='KTM',linewidth=5)
plt.title('bike details in line plot')
plt.ylabel('Distance in kms')
plt.xlabel('Days')
plt.legend()
```

```python
# histogram 

import matplotlib.pyplot as plt


#list of days to plat 
days = [50,80,70,80,40,20,20,20,70,20,60,20,80,50,40,50,20,60,60,60]

#increase bar hist 
bins = [0,10,20,40,50,60,70,80,90,100]

plt.hist(days, bins, histtype='stepfilled', rwidth=0.88)
plt.xlabel('Distance in kms')
plt.ylabel('kilometer count')
plt.title('bike details Histogram')
```
