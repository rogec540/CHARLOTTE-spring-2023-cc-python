```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
restaurant_inventory = {
    'American': 6, 
    'Italian': 8, 
    'French': 2, 
    'Japanese': 4, 
    'Chinese': 3, 
    'Thai': 2,
    'Korean': 1,
    'Mexican': 5,
    'Colombian': 1,
    'Venezuelan': 1,
    'Indian': 2,
    'Fusion': 2,
    'Others': 13,  
    }
```

```python
plt.bar(range(len(restaurant_inventory)), list(restaurant_inventory.values()), align='center', color=['#fdcae1', '#f16d8b', '#d74c22', '#ffb347', '#fdfd96', '#4dc656', '#bffcc6', '#c6e2e9', '#7bd8f1', '#322975', '#b28bff', '#5a008d', '#534d6b'])
plt.xticks(range(len(restaurant_inventory)), list(restaurant_inventory.keys()), rotation = 45)
plt.xlabel('Types of Restaurants')
plt.ylabel('Number of Times on List')
plt.title('Restaurant List Bar Chart')
plt.show()
```

```python
restaurant_types = ['American', 'Italian', 'French', 'Japanese', 'Chinese', 'Thai', 'Korean', 'Mexican', 'Colombian', 'Venezuelan', 'Indian', 'Fusion', 'Other']
restaurant_types.reverse()
num_on_list = [6, 8, 2, 4, 3, 2, 1, 5, 1, 1, 2, 2, 13]
num_on_list.reverse()
```

```python
plt.stem(restaurant_types, num_on_list, orientation = 'horizontal')
plt.xlabel('Number of Times on List')
plt.ylabel('Types of Restaurants')
plt.title('Restaurant List Stem Plot')
plt.show()
```
